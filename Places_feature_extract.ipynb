```python
from keybert import KeyBERT

kw_model=KeyBERT('distilbert-base-nli-mean-tokens')
```

```python
df['desc_list'] = ['' for i in range(len(df))]

for i in range(len(df)):
  keys1 = kw_model.extract_keywords(df['description'][i])
  keys2 = kw_model.extract_keywords(df['reviews_0_text'][i])
  keys3 = kw_model.extract_keywords(df['reviews_1_text'][i])
  all_keys = keys1 + keys2 + keys3
  all_keys = [x[0] for x in keys1] + [x[0] for x in keys2] + [x[0] for x in keys3]
  all_keys = list(set(all_keys))
  df['desc_list'][i] = str(' '.join(all_keys))
```

```python
df['desc_list']
```

```python
import re
import nltk
nltk.download('averaged_perceptron_tagger')


reviews_0=df['reviews_0_text'].values
reviews_1=df['reviews_1_text'].values
desc=df['description'].values
word_tokens_0 = [rev.split() for rev in reviews_0]
word_tokens_1 = [rev.split() for rev in reviews_1]
word_tokens_desc = [rev.split() for rev in desc]

tokens = [[] for i in range(len(df))]
for i in range(len(df)):
  tokens[i] = list(set(word_tokens_0[i] + word_tokens_1[i] + word_tokens_desc[i]))
  for j in range(len(tokens[i])):
    tokens[i][j] = re.sub("[^a-zA-Z]",'',tokens[i][j])
  tokens[i] = [t.lower() for t in tokens[i] if t!='' and t!=' ']

pos_tags=[nltk.pos_tag(rv) for rv in tokens]
adj_list = []
for words in pos_tags:
  adj = []
  for tok in words:
    if tok[1]=='JJ':
      adj.append(tok[0])
  adj = list(set(adj))
  adj = ' '.join(adj)
  adj_list.append(adj)

df['adj_list'] = [adj for adj in adj_list]
print(adj_list)
```

**Review Tags**

```python
cols_to_merge = ['reviewsTags_0_title','reviewsTags_1_title', 'reviewsTags_2_title', 'reviewsTags_3_title', 'reviewsTags_4_title', 'reviewsTags_5_title', 'reviewsTags_6_title', 'reviewsTags_7_title', 'reviewsTags_8_title', 'reviewsTags_9_title']
df['reviewTagsList'] = df.apply(lambda x: ' '.join(set(cel for cel in x.iloc[21:30] if cel != '0' and cel != ' ' and cel != '' and cel != 'nan')), axis=1)

# print the merged column
print(df['reviewTagsList'])
```

```python
for i in range(df.shape[0]):
  list_a = df.iloc[i]['reviewTagsList'].split(' ')
  list_b = df.iloc[i]['desc_list'].split(' ')
  list_c = df.iloc[i]['adj_list'].split(' ')
  set_ = set(list_a + list_b + list_c)
  new_s = {x for x in set_ if x != '' and set != ' ' and set != '0' and set != 'nan'}
  df.at[i, 'feature_set'] = ' '.join(list(new_s))
```

```python
def get_filtered_res(data,qwords):
  sim_score = [[id,0] for id in data['place_id']]
  for i in range(len(data)):
    ft = data['feature_set'][i].split()
    sim_score[i][1] = len(set(ft).intersection(set(qwords)))/len(qwords)
  sim_score.sort(key = lambda x: x[1], reverse = True)
  if len(data) <= 100:
    res_id = [ele[0] for ele in sim_score]
  else:
    res_id = [ele[0] for ele in sim_score[:100]]
  res_df = df.loc[res_id]
  return res_df

res_df = get_filtered_res(df,["hill","himalaya","beaautiful","trekking","peaceful", "mountain", "himachal"])
res_df
```
